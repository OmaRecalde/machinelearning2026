#### 2.plot()
Múltiples parámetros:
`plt.plot(x, y, style, label, linewidth=1, alpha=1)`

* `'style'`: Cadena de texto con formato (color, tipo...) `[color][marker][line]`
* `'label'`: Referencia para leyenda
* `'linewidth'`: Ancho de línea / brocha
* `'alpha'`: Transparencia

```python
y_data = np.random.randint(100, size=(10))
x_data = np.arange(10)

fig, axis = plt.subplots(1, 1)
axis.plot(x_data, y_data, 'gh--', label='dollars', linewidth=1, alpha=0.6) # creación del gráfico
axis.legend(loc='best') # representar la leyenda

plt.show()
```

# 6 frame.plot()
`frame.plot(x, y, label=<Series>, ax, style, kind, xticks, yticks, title, subplots)`

* **(solo para Series) label:** Referencia para la leyenda
* **ax:** Subfigura en la que dibujar los datos
* **style:** Estilo de la línea
* **kind:** Tipo de gráfico (`'bar'`, `'pie'`, `'hist'`, `'area'`, `'line'`, `'barh'`, `'density'`, `'kde'`).
* **xticks y yticks:** Valores en los ejes X e Y
* **title:** String como título
* **(solo para DataFrame) subplots:** Bool si se desean subfiguras separadas para cada columna.
* **(solo para DataFrame) 'x' e 'y':** Se pueden utilizar para seleccionar columnas para cada eje

[Lista completa de argumentos](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html)

```python
rand_matrix = np.random.randint(100, size=(10, 3))
frame = pd.DataFrame(rand_matrix, columns=list('ABC'))

frame.plot(style=['r-', 'g--', 'b:'], xticks=range(-2, 12), title='Random lines', subplots=True)
plt.show()

display(frame)
```

## 7.1 Relaciones entre variables
`sns.relplot()`

```python
tips = sns.load_dataset("tips")
display(tips.sample(5))

# Scatter relplot
sns.relplot(x='total_bill', y='tip', hue='sex', data=tips, kind='scatter')
plt.show()

# representar la media y la variación de una variable en función de otra
data = sns.load_dataset('flights')
sns.relplot(x="year", y="passengers", kind="line", data=data)
plt.show()

# relacionar hasta 4 variables automáticamente
sns.relplot(x="total_bill", y="tip", hue="smoker", col="sex", height=4, kind="scatter", data=tips) # kind ('line', 'scatter'...)
plt.show()
```

## 7.2 Distribución de un dataset
* **kde:** gaussian kernel density estimate, línea de densidad
* **bins:** tamaño de los contenedores
* **rug:** densidad de los datos

```python
# histograma y density plot
sns.displot(tips['tip'], kde=True, bins=10, rug=True)
plt.show()

# Distribución bi-variable
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()

# Density plot bi-variable
iris = sns.load_dataset('iris')
display(iris.head(5))
display(iris["species"].unique())

sns.jointplot(x='petal_length', y='petal_width', kind="kde", data=iris)
plt.show()

# Relación de cada pareja
sns.pairplot(iris)
plt.show()

# Relación de cada pareja coloreado por clase
sns.pairplot(iris, hue="species")
plt.show()
```

```python
# scatter de clases, coloreando segun clase en matplotlib
iris = sns.load_dataset('iris')

# preparacion de los datos
data = []
colors = ["red", "green", "blue"]
groups = ["setosa", "versicolor", "virginica"]
columns = ["petal_length", "petal_width"]

for i in groups:
    # busqueda por clase definida
    data.append(iris[iris["species"] == i][columns].values)

# creacion de la grafica
fig = plt.figure()
ax = fig.add_subplot()

for count, d in enumerate(data):
    # separar las variables del grafico
    x, y = d[:, 0], d[:, 1]
    ax.scatter(x, y, alpha=0.4, c=colors[count - 1], edgecolors='none', s=30, label=groups[count - 1])

# etiquetas
ax.set_xlabel(columns[0])
ax.set_ylabel(columns[1])
plt.legend(loc='best')
plt.title('Matplot scatter plot')
plt.show()
```

## 7.3 Visualización de correlaciones
* Regresión lineal con variables
* Correlaciones numéricas

```python
# correlación entre cuenta total y propina, recta regresion e intervalo de confianza
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()

# correlación con valores categóricos (variable discreta, variable debe ser numérica/continua) not a good idea
sns.regplot(x="size", y="tip", data=tips)
plt.show()

# regresión con polinomios de más grados
sns.regplot(x="petal_length", y="sepal_width", data=iris, order=5)
plt.show()

# Correlaciones con varias variables
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, col='time', order=2)
plt.show()
```

* Carga los datos de un dataset de seaborn (distinto a 'tips', 'iris' o 'flights'): https://github.com/mwaskom/seaborn-data
* Si para alguno de los ejercicios no encuentras ejemplos en el dataset, busca otro para completarlo.

* Representa visualmente la relación de cada pareja de atributos (variables).

* Representa los valores de dos variables (x e y, numéricas) en base a otra variable (categórica).
* Crea los gráficos que demuestren lo siguiente en el dataset seleccionado:
  * La distribución de valores (media, distribución, outliers) de una variable numérica continua.
  * Compara la distribución de esa variable con otra (continua y numérica) para representar cuantitativamente la densidad de elementos por valor en ambas variables.
  * Muestra un ejemplo en el dataset de una variable a la que un modelo de regresión lineal de orden > 1 sea más ajustado que uno de orden == 1.

```python
# --- Solución 8.1 Matplotlib ---

# 1. Crear la lista de años de 2000 a 2020
years = np.arange(2000, 2021)
n_points = len(years)

# Semilla para reproductibilidad
np.random.seed(42)

# 2. Generar datos aleatorios (4 secuencias distintas para simulated GDP)
y0 = np.random.randint(100, 500, size=n_points) # Hawaii
y1 = np.random.randint(50, 300, size=n_points)  # San Marino
y2 = np.random.randint(80, 400, size=n_points)  # Islas Feroe
y3 = np.random.randint(120, 600, size=n_points) # Guayana

labels = ["Hawaii", "San Marino", "Islas Feroe", "Guayana"]
series = [y0, y1, y2, y3]
colors = ['blue', 'orange', 'green', 'red']

# 3. Representar las 4 secuencias en una sola figura
fig, ax = plt.subplots(figsize=(10, 5))
for data, label, color in zip(series, labels, colors):
    ax.plot(years, data, label=label, color=color, marker='o')

ax.set_title("Datos aleatorios - Comparativa Global")
ax.set_xlabel("Año")
ax.set_ylabel("GDP")
ax.legend(loc='upper left')
plt.grid(True)
plt.show()

# 4. Representar los mismos datos en 4 subfiguras (2x2)
fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)
fig.suptitle("Datos aleatorios", fontsize=16)

axes_flat = axes.flatten()

for i, ax in enumerate(axes_flat):
    ax.plot(years, series[i], label=labels[i], color=colors[i], marker='s')
    ax.set_xlabel("Año")
    ax.set_ylabel("GDP")
    ax.legend(loc='upper left')
    ax.grid(True)

# 5. Añadir anotación al primer gráfico (Hawaii)
max_idx = np.argmax(y0)
max_year = years[max_idx]
max_val = y0[max_idx]

axes_flat[0].annotate(f'Pico Máximo ({max_val})', 
                     xy=(max_year, max_val), 
                     xytext=(max_year - 5, max_val + 30),
                     arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6))

plt.tight_layout()
plt.show()
```

```python
# --- Solución 8.2 Seaborn ---

# 1. Cargar dataset 'penguins' (distinto a tips, iris o flights)
penguins = sns.load_dataset("penguins")
display(penguins.head())

# Limpiar valores nulos para evitar inconsistencias
penguins_clean = penguins.dropna()

# 2. Representar visualmente la relación de cada pareja de atributos (pairplot)
sns.pairplot(penguins_clean, hue="species")
plt.suptitle("Relación entre parejas de variables - Penguins", y=1.02)
plt.show()

# 3. Pareja de variables continuas y demostración de su correlación (flipper_length vs body_mass)
sns.regplot(x="flipper_length_mm", y="body_mass_g", data=penguins_clean, color="teal")
plt.title("Correlación: Largo de Aleta vs Masa Corporal")
plt.xlabel("Largo de Aleta (mm)")
plt.ylabel("Masa Corporal (g)")
plt.show()

# 4. Representar x e y (numéricas) en base a variable categórica (bill_length vs bill_depth por species)
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", hue="species", style="sex", data=penguins_clean, s=70)
plt.title("Largo vs Profundidad del Pico por Especie y Sexo")
plt.xlabel("Largo del Pico (mm)")
plt.ylabel("Profundidad del Pico (mm)")
plt.show()

# 5. Distribución de valores (media, distribución, outliers) con Boxplot
sns.boxplot(x="species", y="body_mass_g", data=penguins_clean, palette="Set2")
plt.title("Distribución de Masa Corporal por Especie (Boxplot)")
plt.xlabel("Especie")
plt.ylabel("Masa Corporal (g)")
plt.show()

# 6. Comparar densidad de dos variables continuas (jointplot KDE bi-variable)
sns.jointplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_clean, kind="kde", fill=True, cmap="Blues")
plt.suptitle("Densidad Conjunta de Dimensiones del Pico", y=1.02)
plt.show()

# 7. Modelo de regresión lineal de orden > 1 vs orden == 1
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Orden = 1
sns.regplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_clean, ax=axes[0], order=1, line_kws={"color": "red"})
axes[0].set_title("Regresión Lineal Orden = 1")

# Orden = 2 (Polinómica)
sns.regplot(x="bill_length_mm", y="bill_depth_mm", data=penguins_clean, ax=axes[1], order=2, line_kws={"color": "green"})
axes[1].set_title("Regresión Polinómica Orden = 2")

plt.tight_layout()
plt.show()
```
