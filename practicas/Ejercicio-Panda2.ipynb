## Sección 1: Búsqueda, Verificación de Valores y Conteo de Elementos

En esta sección aprenderemos a auditar un `DataFrame` para verificar la existencia de variables específicas, mapear la presencia de un listado de valores usando máscaras booleanas y cuantificar el total de registros que cumplen con condiciones de umbral.

### 📌 Conceptos Clave:
* **Operador `in`:** Valida de forma nativa si una etiqueta o nombre de columna se encuentra registrado en el esquema de claves del `DataFrame`.
* **Método `.isin([valores])`:** Genera una matriz booleana (máscara) indicando qué coordenadas contienen elementos que coinciden con la lista suministrada.
* **Capa subyacente NumPy:** Al encadenar `.values`, se extrae el array primitivo de NumPy subyacente en Pandas, facilitando operaciones vectorizadas veloces como sumatorias totales (`.sum()`).

```python
import numpy as np
import pandas as pd

# =====================================================================
# 1. CONFIGURACIÓN INICIAL Y GENERACIÓN DE MATRIZ ALEATORIA
# =====================================================================
print(f"[INFO] Versión de Pandas en uso: {pd.__version__}")

# Fijamos una semilla aleatoria interna para reproducibilidad en tus pruebas
np.random.seed(42)

# Generamos una matriz aleatoria de enteros entre 0 y 5 con dimensiones 2x3
rand_matrix = np.random.randint(6, size=(2, 3))
frame = pd.DataFrame(rand_matrix, columns=list('ABC'))

print("\n--- DataFrame Base de Búsqueda ---")
display(frame)


# =====================================================================
# 2. BÚSQUEDA DE COLUMNAS (VERIFICACIÓN DE ESQUEMA)
# =====================================================================
# El DataFrame evalúa por defecto sobre sus ejes de columnas (claves)
existe_columna = 'A' in frame
print(f"\n¿Existe la columna 'A' en el DataFrame?: {existe_columna}")


# =====================================================================
# 3. VERIFICACIÓN DE VALORES ESPECÍFICOS CON .isin()
# =====================================================================
# Mapeamos la presencia de los números 2 o 3 dentro de toda la estructura
mascara_presencia = frame.isin([3, 2])
print("\n--- Máscara Booleana de Presencia (.isin([3, 2])) ---")
display(mascara_presencia)


# =====================================================================
# 4. CUANTIFICACIÓN COMPLETA DE OCURRENCIAS
# =====================================================================
# Contar exactamente cuántos elementos en el DataFrame son iguales a 4
total_cuatros = frame.isin([4]).values.sum()
print(f"\nCantidad total de valores '4' encontrados: {total_cuatros}")

# Evaluar condiciones relacionales numéricas combinadas con sumatorias
# ¿Cuántos valores son mayores o iguales a 2?
mask_umbral = frame >= 2
total_mayores_dos = mask_umbral.values.sum()

print("\n--- Criterio de Evaluación Relacional (>= 2) ---")
print(f"Total de celdas que cumplen la condición: {total_mayores_dos}")
display(mask_umbral)
```

## Sección 2: Técnicas de Ordenación (Sorting) y Ranking de Datos

En este bloque analizaremos cómo reordenar las estructuras tabulares bajo criterios jerárquicos y cómo establecer rankings numéricos basados en los valores de las variables.

### 📌 Conceptos Clave:
* **`.sort_index()`**: Reordena las filas o columnas según sus etiquetas de índice o nombres. El parámetro `axis=1` se usa para ordenar alfabéticamente las columnas[cite: 3].
* **`.sort_values(by=...)`**: Ordena las observaciones basándose en los valores reales de una o más columnas[cite: 3]. Si se pasa `axis=1`, ordena las columnas de acuerdo con los valores de una fila específica[cite: 3].
* **`inplace=True`**: Modifica la estructura de forma definitiva en memoria sin necesidad de reasignar la variable[cite: 3].
* **`.rank(method='max')`**: Asigna posiciones numéricas (rankings) a los elementos[cite: 3]. El método `'max'` gestiona empates asignando la posición máxima del grupo repetido[cite: 3].

```python
from random import shuffle
import numpy as np
import pandas as pd

# =====================================================================
# 1. CREACIÓN DEL DATAFRAME CON ÍNDICES MEZCLADOS
# =====================================================================
np.random.seed(10)
rand_matrix = np.random.randint(20, size=(5, 4))

# Generamos una lista de índices y la mezclamos aleatoriamente
indices = list(range(5))
shuffle(indices)

# Creamos el DataFrame con columnas desordenadas alfabéticamente
df_desordenado = pd.DataFrame(rand_matrix, columns=list('DACB'), index=indices)
print("--- DataFrame Original (Índices y Columnas Desordenadas) ---")
display(df_desordenado)


# =====================================================================
# 2. ORDENACIÓN POR ÍNDICES Y EJES
# =====================================================================

# Ordenar las filas por su índice semántico en orden descendente (De mayor a menor)
print("\n--- Ordenado por Índice de Filas Descendente (.sort_index) ---")
display(df_desordenado.sort_index(ascending=False))

# Ordenar las columnas alfabéticamente (axis=1) en orden descendente (D -> C -> B -> A)
print("\n--- Ordenado por Eje de Columnas Descendente (axis=1) ---")
display(df_desordenado.sort_index(axis=1, ascending=False))


# =====================================================================
# 3. ORDENACIÓN POR VALORES CONCRETOS
# =====================================================================

# Ordenar las filas de menor a mayor tomando como referencia los valores de la columna 'A'
print("\n--- Filas Ordenadas Ascendentemente por los valores de la columna 'A' ---")
display(df_desordenado.sort_values(by='A', ascending=True))

# Aplicar ordenación persistente en memoria usando inplace=True
df_desordenado.sort_values(by='A', ascending=False, inplace=True)
print("\n--- DataFrame Modificado en Memoria (Orden Descendente por 'A') ---")
display(df_desordenado)


# =====================================================================
# 4. RANKING E ITERACIÓN DE VALORES ORDERADOS
# =====================================================================

# Construir un ranking posicional por filas (axis=1) con método de resolución 'max'
print("\n--- Ranking de Valores por Filas (.rank) ---")
display(df_desordenado.rank(method='max', axis=1))

# Imprimir los valores de la columna 'C' ordenados de mayor a menor uno por uno
print("\n--- Bucle: Valores de la columna 'C' ordenados de Mayor a Menor ---")
for valor in df_desordenado.sort_values(by='C', ascending=False)['C'].values:
    print(f"Valor extraído de C: {valor}")
```

## Sección 3: Operaciones Aritméticas, Alineación de Índices y Difusión (Broadcasting)

En esta sección analizaremos cómo interactúan dos matrices de distintas dimensiones al realizar operaciones matemáticas estándar y cómo usar métodos explícitos para controlar la introducción de valores nulos (NaN).

### 📌 Conceptos Clave:
* **Alineación por Índices:** Al sumar, restar o multiplicar DataFrames, Pandas empareja las filas y columnas que tengan exactamente el mismo nombre. Si no coinciden, introduce de forma predeterminada un `NaN`.
* **Parámetro `fill_value`:** Permite especificar un valor por defecto (como `0`) para rescatar los datos en caso de que una de las celdas no exista en la otra matriz[cite: 3].
* **Broadcasting (Difusión):** Operación matemática que permite restar o sumar una sola fila (`Series`) a todo el `DataFrame` completo, repitiendo la operación renglón por renglón de manera automatizada[cite: 3].

```python
import numpy as np
import pandas as pd

# =====================================================================
# 1. OPERACIONES MATEMÁTICAS ESTÁNDAR ENTRE DATAFRAMES SIMÉTRICOS
# =====================================================================
np.random.seed(20)

matrixA = np.random.randint(100, size=(4, 4))
matrixB = np.random.randint(100, size=(4, 4))

frameA = pd.DataFrame(matrixA)
frameB = pd.DataFrame(matrixB)

print("--- DataFrame A ---")
display(frameA)
print("\n--- DataFrame B ---")
display(frameB)

# Suma mediante operadores aritméticos convencionales (+) y métodos explícitos (.add)
print("\n--- Resultado de la Suma (Sintaxis Equivalentes) ---")
display(frameA + frameB)

# Resta mediante operadores aritméticos convencionales (-) y métodos explícitos (.sub)
print("\n--- Resultado de la Resta (frameB - frameA) ---")
display(frameB.sub(frameA))


# =====================================================================
# 2. ALINEACIÓN ASIMÉTRICA Y USO DE FILL_VALUE
# =====================================================================
# Creamos una tercera matriz más pequeña (dimensiones 3x3)
frameC = pd.DataFrame(np.random.randint(100, size=(3, 3)))

print("\n--- DataFrame C (Asimétrico 3x3) ---")
display(frameC)

# La suma directa generará valores NaN en la cuarta fila y cuarta columna (índices 3)
print("\n--- Suma Incompleta Directa (Aparición de NaNs) ---")
display(frameC + frameA)

# SOLUCIÓN DE CONTROL: Usar .add() especificando fill_value=0 para rescatar los datos huérfanos
print("\n--- Suma Controlada de forma Segura (.add con fill_value=0) ---")
display(frameA.add(frameC, fill_value=0))


# =====================================================================
# 3. OPERACIONES ENTRE SERIES Y DATAFRAMES (BROADCASTING)
# =====================================================================
# Uso común en analítica: Averiguar la diferencia o desviación de cada fila frente a un patrón
rand_matrix = np.random.randint(10, size=(3, 4))
df = pd.DataFrame(rand_matrix, columns=list('ABCD'))

print("\n--- DataFrame Base para Difusión ---")
display(df)

# Extraemos la primera fila (Índice posicional 0), la cual se convierte en una Series
primera_fila = df.iloc[0]
print("\n--- Fila de Referencia (Series de la fila 0) ---")
display(primera_fila)

# Restar la primera fila a todas las filas del DataFrame (Broadcasting por filas)
print("\n--- Desviación del DataFrame con respecto a la Fila 0 (Eje Horizontal) ---")
display(df - primera_fila)

# Restar una columna completa a todas las columnas usando .sub() especificando axis=0
print("\n--- Desviación con respecto a la columna 'A' (Eje Vertical, axis=0) ---")
display(df.sub(df['A'], axis=0))
```

## Sección 4: Funciones Lambda a Medida y Estadísticas Descriptivas

En esta sección estudiaremos cómo vectorizar transformaciones personalizadas sobre los ejes de una matriz y cómo perfilar un conjunto de datos extrayendo métricas estadísticas agregadas.

### 📌 Conceptos Clave:
* **`.apply(lambda)`:** Aplica una función anónima (o personalizada) a lo largo de un eje del DataFrame.
  * `axis=0` (Por defecto): Procesa los datos columna por columna[cite: 3].
  * `axis=1`: Procesa los datos fila por fila[cite: 3].
* **`idxmax()` / `idxmin()`:** Devuelven el **índice semántico** (la etiqueta) donde se localiza el valor máximo o mínimo absoluto de una variable[cite: 3].
* **`numeric_only=True`:** Restringe los cálculos aritméticos y estadísticos exclusivamente a columnas numéricas, ignorando automáticamente variables de texto (como nombres o géneros) para evitar errores sintácticos[cite: 3].

```python
import numpy as np
import pandas as pd

# =====================================================================
# 1. APLICACIÓN DE FUNCIONES PERSONALIZADAS (LAMBDA)
# =====================================================================
np.random.seed(30)
rand_matrix = np.random.randint(10, size=(3, 4))
df_lambda = pd.DataFrame(rand_matrix, columns=list('ABCD'))

print("--- DataFrame Base para Funciones Lambda ---")
display(df_lambda)

# Ejemplo 1: Calcular el rango (Máximo - Mínimo) de cada FILA usando axis=1
rango_filas = df_lambda.apply(lambda x: x.max() - x.min(), axis=1)
print("\n--- Rango numérico calculado por cada FILA (axis=1) ---")
print(rango_filas)

# Ejemplo 2: Calcular el rango usando una función explícita por COLUMNA usando axis=0
def calcular_rango_columna(columna):
    return columna.max() - columna.min()

rango_columnas = df_lambda.apply(calcular_rango_columna, axis=0)
print("\n--- Rango numérico calculado por cada COLUMNA (axis=0) ---")
print(rango_columnas)


# =====================================================================
# 2. ESTADÍSTICA DESCRIPTIVA EN ACCIÓN
# =====================================================================
# Construcción de un DataFrame heterogéneo (Texto, Números, Nulos)
diccionario = { 
    "nombre": ["Marisa", "Laura", "Manuel", "Carlos"], 
    "edad": [34, 34, 11, 30],
    "puntos": [98, 12, 98, np.nan], 
    "genero": ["F", "F", "M", "M"]
}
frame_estudiantes = pd.DataFrame(diccionario)

print("\n--- DataFrame de Estudiantes ---")
display(frame_estudiantes)

# .describe() genera un resumen estadístico instantáneo de las variables numéricas
print("\n--- Resumen Estadístico Estándar (.describe()) ---")
display(frame_estudiantes.describe())

# Sumatoria vertical restringida a datos numéricos (numeric_only=True)
print("\n--- Sumatoria total de las columnas numéricas ---")
print(frame_estudiantes.sum(axis=0, numeric_only=True))

# Desviación estándar de una columna específica
desviacion_edad = frame_estudiantes['edad'].std()
print(f"\nDesviación estándar de la variable 'edad': {desviacion_edad:.2f}")


# =====================================================================
# 3. BÚSQUEDA FILTRADA POR ÍNDICES CRÍTICOS (MÁXIMOS Y MÍNIMOS)
# =====================================================================

# Identificar el índice semántico de la persona con la edad más alta
indice_max_edad = frame_estudiantes['edad'].idxmax()
print(f"\nÍndice semántico del estudiante con mayor edad: {indice_max_edad} (Fila {indice_max_edad})")

# EXTRAER LA FILA COMPLETA que contenga el valor máximo de una columna
valor_max_puntos = frame_estudiantes['puntos'].max()
print(f"\nPuntuación máxima registrada: {valor_max_puntos}")

print("\n--- Registro(s) del estudiante con el puntaje máximo obtenido ---")
# Filtramos buscando las filas que coincidan exactamente con dicho valor máximo
filas_max_puntaje = frame_estudiantes[frame_estudiantes['puntos'] == valor_max_puntos]
display(filas_max_puntaje)
```

## Sección 5: Agregaciones (`groupby`), Filtrado Colectivo y Matriz de Correlación

En esta sección estudiaremos cómo agrupar información categórica, extraer resúmenes por segmentos y calcular la correlación lineal entre variables para identificar dependencias numéricas.

### 📌 Conceptos Clave:
* **`.groupby('columna')`:** Segmenta el DataFrame en grupos basados en los valores únicos de la columna especificada.
* **`.aggregate(['min', 'mean', 'max'])`:** Permite calcular múltiples estadísticos descriptivos en una sola línea de código sobre las columnas seleccionadas.
* **`.filter(funcion)`:** Evalúa una condición lógica sobre todo el grupo completo (por ejemplo, si la media del grupo supera un umbral) y descarta los registros de los grupos que no la cumplan[cite: 3].
* **`.corr(numeric_only=True)`:** Devuelve una matriz que mide el grado de relación lineal entre las variables numéricas (valores entre -1 y 1)[cite: 3].

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Recuperamos el DataFrame de estudiantes de la sección anterior
diccionario = { 
    "nombre": ["Marisa", "Laura", "Manuel", "Carlos"], 
    "edad": [34, 34, 11, 30],
    "puntos": [98, 12, 98, np.nan], 
    "genero": ["F", "F", "M", "M"]
}
frame_estudiantes = pd.DataFrame(diccionario)


# =====================================================================
# 1. AGREGACIONES AGRUPADAS CON GROUPBY
# =====================================================================

print("--- Conteo de Registros por Género ---")
# Agrupamos por género y contamos los elementos no nulos
display(frame_estudiantes.groupby('genero').count())

print("\n--- Múltiples Estadísticos Simultáneos (.aggregate) ---")
# Agrupamos por género y obtenemos min, media y max de edad y puntos
estadisticas_grupo = frame_estudiantes.groupby('genero')[['edad', 'puntos']].aggregate(['min', 'mean', 'max'])
display(estadisticas_grupo)


# =====================================================================
# 2. FILTRADO LÓGICO DE GRUPOS COMPLETOS
# =====================================================================

# Definimos una función que evalúa si la media de edad de un grupo es mayor a 30
def evaluar_media_edad(grupo):
    return grupo["edad"].mean() > 30

print("\n--- Filtrado de Grupos (Solo géneros con promedio de edad > 30) ---")
df_filtrado_grupo = frame_estudiantes.groupby('genero').filter(evaluar_media_edad)
display(df_filtrado_grupo)


# =====================================================================
# 3. CASO REAL: ANÁLISIS DE CORRELACIONES (AUTO MPG DATASET)
# =====================================================================

# Ingesta remota segura del dataset de eficiencia de combustible de vehículos
path_url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
mpg_data = pd.read_csv(
    path_url, 
    sep='\s+', 
    header=None,
    names=['mpg', 'cilindros', 'desplazamiento', 'potencia', 'peso', 'aceleracion', 'año', 'origen', 'nombre'],
    na_values='?', 
    engine='c'
)

print("\n--- Muestra Aleatoria de la Eficiencia de Motores (Auto MPG) ---")
display(mpg_data.sample(3))

print("\n--- Matriz Completa de Correlación Numérica ---")
# Eliminamos columnas contextuales ('año', 'origen') para enfocar el análisis matemático
matriz_corr = mpg_data.drop(['año', 'origen'], axis=1).corr(numeric_only=True)
display(matriz_corr)

# Aplicamos un gradiente de color (estilo mapa de calor) para identificar correlaciones visualmente
print("\n[INFO] Desplegando representación gráfica de la matriz de correlación...")
display(matriz_corr.style.background_gradient(cmap=plt.get_cmap('RdYlGn'), axis=1))


# =====================================================================
# 🌟 EJERCICIO MODELO (PLANTILLA TIPO EXAMEN)
# =====================================================================
# Problema representativo de "100 Pandas Puzzles": 
# "Dado un DataFrame numérico, restar la media de cada fila a cada elemento de dicha fila"
print("\n--- 📝 EJERCICIO MODELO: Normalización por Resta de Media de Filas ---")

np.random.seed(42)
df_ejercicio = pd.DataFrame(np.random.randint(1, 10, size=(3, 3)), columns=list('XYZ'))
print("DataFrame Original del Ejercicio:")
display(df_ejercicio)

# Calculamos el promedio horizontal de cada fila (axis=1)
media_filas = df_ejercicio.mean(axis=1)

# Usamos el método .sub indicando el eje vertical axis=0 para aplicar la resta fila por fila
df_normalizado = df_ejercicio.sub(media_filas, axis=0)
print("\nDataFrame Resultante (Cada celda restada el promedio de su propia fila):")
display(df_normalizado)
```
