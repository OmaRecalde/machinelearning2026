## 3. Iteración (bucles)

- Repetición de un bloque de código.

### Bucle `while`

- Repetición de un bloque de código hasta que se deje de cumplir una expresión (es decir, hasta que una condición evalúe a `False`).
- Si la condición evalúa a `False` desde el principio, el bloque de código nunca se ejecuta.
- Cuidado con los bucles infinitos.

Formato general:
```python
while test:      # Mientras se cumple la condición
    statements    # Instrucciones a ejecutar
```

````python
# Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]
while indice < 3:
    print(numeros[indice])
    indice = indice + 1
````

````python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0
while indice < len(numeros):
    # Reviso cada número de la lista; si es menor a 10 lo muestro y sumo al contador
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1
print('Contador:', contador)
print('Indice:', indice)
````

````python
nombre = 'Pablo'
while nombre:  # Mientras 'nombre' no sea vacío
    print(nombre)
    nombre = nombre[1:]
````

````python
# Aunque el comentario original dice "bucle infinito", en realidad SÍ termina
# porque la condición i < 10 eventualmente se vuelve False
i = 0
while i < 10:
    print(i)
    i += 1
````

### 3.1 Las sentencias break, continue y else

**Break y continue**
- Solo tienen sentido dentro de bucles.
- `break` permite terminar el bucle por completo.
- `continue` permite saltar a la siguiente iteración, continuando con el bucle.
- Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sentencias condicionales (`if`).

````python
# Ejemplo: break
# Busco un rating específico en la lista; cuando lo encuentro, muestro "Found" y detengo el bucle
rating_to_find = 4.2
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]
for rating in movie_ratings:
    print(rating)
    if rating == rating_to_find:
        print("Found")
        break
print(rating)
````

````python
for i in range(10):
    if i % 2 == 0:
        continue  # Si el número es par, salta a la siguiente iteración
    print(f'Odd number {i}')
````

Observa cómo la sentencia `continue` te puede ayudar a reducir el número de niveles de anidamiento. Sin `continue`, el anterior ejemplo sería:

````python
for i in range(10):
    if i % 2 != 0:
        print('Numero impar:', end=' ')
        print(i)
````

````python
# Nota: el 'break' aquí solo detiene el bucle interno (el de 'a'), no el externo (el de 'i')
for i in range(5):
    for a in range(2):
        print(f"{i} - {a}")
        if i == 3 and a == 0:
            break
````

### Else

- Los bucles pueden tener una sentencia `else`.
- Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros lenguajes.
- Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de un `break`.

````python
lista = [60, 60, 30, 60]
element_to_find = 60
for element in lista:
    if element == element_to_find:
        print("found")
        break
else:
    print("not found")
````

````python
# Misma lógica que la celda anterior, pero usando una variable "found" como bandera
# en vez de aprovechar el 'else' del bucle for
lista = [10, 30, 50, 40]
element_to_find = 30
found = False
for element in lista:
    if element == element_to_find:
        found = True
        break
if found:
    print("dato encontrado")
else:
    print("not found")
````

## 4. List Comprehensions

- Permiten construir listas a través de la ejecución repetida (sentencia `for`) de una expresión para cada item de un objeto iterable.
- Van entre `[` y `]`. Esto es indicativo de que estamos construyendo una lista.

Sintaxis:
```python
[<expression> for <item> in <iterable>]
```

Ejemplo: repetir los caracteres de un string.

````python
[char*3 for char in "Enrique"]
````

````python
lista = []
for char in 'Enrique':
    lista.append(char)
print(lista)
````

Comúnmente, el item de la sentencia `for` aparecerá en la expresión principal, pero eso no es obligatorio.

````python
[2 for _ in 'Enrique']
````

### 4.0.1 Versión extendida

Se puede especificar un filtro (sentencia `if`) para obtener únicamente los elementos que cumplan cierta condición.

Sintaxis:
```python
[<expression> for <item> in <iterable> if <condition>]
```

Ejemplo: obtener números pares:

````python
[y for y in range(9) if y % 2 == 0]
````

Las list comprehensions no son realmente requeridas, ya que siempre podemos escribir un bucle equivalente.

````python
pares = []
for x in range(9):
    if x % 2 == 0:
        pares.append(x)
print(pares)
````

### 4.0.2 Versión completa

La sentencia `if` de una list comprehension también puede contener una expresión alternativa.

Sintaxis:
```python
[<expression_1> if <condition> else <expression_2> for <item> in <iterable>]
```

Ejemplo: poner a cero los números pares.

````python
[0 if x % 2 == 0 else x for x in range(9)]
````

````python
# Ojo: aquí hay dos "if" distintos.
# El primero (x**2 if x > 2 else x) es parte de la expresión: decide QUÉ valor poner.
# El segundo (if x > 0, al final) es el filtro: decide SI el elemento entra en la lista.
[x**2 if x > 2 else x for x in range(-4, 5) if x > 0]
````

````python
lista = []
for x in range(-4, 5):
    if x > 0:
        if x > 2:
            lista.append(x**2)
        else:
            lista.append(x)
print(lista)
````

````python
[x**2 if x > 0 else 100 if x == 0 else x for x in range(-4, 5)]
````

````python
lista = []
for x in range(-4, 5):
    if x > 0:
        lista.append(x**2)
    elif x == 0:
        lista.append(100)
    else:
        lista.append(x)
print(lista)
````

### 4.0.3 Anidamiento

Las list comprehensions soportan anidamiento en sus expresiones.

Ejemplo: bucle anidado para obtener una lista de tuplas que combinan los elementos de dos listas dadas.

````python
lista_1 = [1, 2, 3]
lista_2 = [4, 5]
[(x, y) for x in lista_1 for y in lista_2]
````

### 4.0.4 Pros y contras

Principales ventajas de las comprehensions en comparación a un bucle convencional:
- Expresión compacta y legible, si estás familiarizado con la sintaxis.
- Mejor rendimiento.

Desventaja: no escalan bien. Una list comprehension se puede convertir rápidamente en una expresión difícil de entender.

### 4.0.5 Otros tipos de comprehensions

**Dictionary comprehensions**

Usando `{` y `}` como delimitadores y una expresión `clave : valor`, se obtiene un diccionario en lugar de una lista.

Ejemplo: diccionario donde cada valor es el cuadrado de la clave.

````python
d = {x: x*x for x in range(10)}
print(d)
print(type(d))
````

````python
# zip combina dos listas elemento por elemento (par a par) en tuplas,
# aquí se usa para armar un diccionario producto -> precio
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
shopping = {k: v for k, v in zip(items, price)}
print(shopping)
````

````python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
for k in zip(items, price):
    print(type(k))
    print(k)
````

## 5. Excepciones

- Las excepciones son eventos que representan situaciones excepcionales.
- Alteran el flujo de ejecución convencional.
- Python lanza excepciones automáticamente cuando se producen errores.
- El programador puede lanzar excepciones de manera explícita (`raise`) y también capturar excepciones para actuar como se crea conveniente.

**Try/except:** permite capturar excepciones y actuar en consecuencia (por ejemplo, ante un `IndexError` al acceder fuera de rango de una lista).

**Try/except genérico:** se puede usar un `except` sin especificar tipo para capturar cualquier error, aunque lo recomendable es ser lo más específico posible (ej. `except IndexError`).

**Capturar el mensaje de error:** con `except Exception as e`, se puede acceder al mensaje de la excepción (`str(e)`) para mostrarlo por pantalla.

**Try/finally:** el bloque `finally` se ejecuta siempre, ocurra o no una excepción. Se usa comúnmente para liberar recursos.

**raise:** permite lanzar excepciones de manera explícita (por ejemplo, `raise IndexError('mensaje')`).
