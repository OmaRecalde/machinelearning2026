Sección 3: Iteración, Uso como Diccionarios y Operaciones entre SeriesEn esta sección avanzamos hacia la manipulación estructurada de las Series. Veremos cómo recorrerlas eficientemente, cómo Pandas nos permite tratarlas con una sintaxis idéntica a la de los diccionarios nativos de Python (añadiendo, eliminando o validando claves) y, lo más importante, cómo se alinean automáticamente los datos cuando realizamos operaciones matemáticas entre dos series independientes.  Nota técnica importante: Cuando sumas o restas dos series en Pandas, la operación no se hace por posición física, sino por coincidencia de índices semánticos. Si un índice existe en una serie pero no en la otra, Pandas no puede adivinar el resultado y por defecto asignará un valor nulo (NaN). Aprenderemos a corregir este comportamiento con máscaras de filtrado.  

```python
import numpy as np
import pandas as pd

# =====================================================================
# 1. ITERACIÓN SOBRE SERIES
# =====================================================================
# Usaremos una serie basada en la secuencia de Fibonacci para las pruebas
fibonacci = pd.Series([0, 1, 1, 2, 3, 5, 8, 13, 21])

print("--- Iteración sobre VALORES ---")
for value in fibonacci:
    print(f"Valor: {value}")

print("\n--- Iteración sobre ÍNDICES ---")
for idx in fibonacci.index:
    print(f"Índice: {idx}")

print("\n--- Iteración SIMULTÁNEA (Índice y Valor con .items()) ---")
# .items() es la forma estándar y más eficiente en Pandas para desempaquetar ambos ejes
for idx, value in fibonacci.items():
    print(f"Índice: {idx} -> Valor: {value}")


# =====================================================================
# 2. SERIES TRATADAS COMO DICCIONARIOS
# =====================================================================
# Podemos inicializar una serie directamente pasando un diccionario de Python
notas_dict = {'Carlos': 100, 'Marcos': 98}
serie_notas = pd.Series(notas_dict)

print("\n--- Serie creada desde Diccionario ---")
print(serie_notas)

# Validar si una clave existe en el índice de la serie (operación 'in')
if 'Marcos' in serie_notas:
    print(f"\nMarcos está en la serie. Su nota es: {serie_notas['Marcos']}")

# Añadir un nuevo elemento o modificar uno existente mediante su clave
serie_notas['Pedro'] = 15
print("\n--- Tras añadir a Pedro ---")
print(serie_notas)

# Eliminar un elemento usando la sentencia nativa 'del'
del serie_notas['Marcos']
print("\n--- Tras eliminar a Marcos ---")
print(serie_notas)


# =====================================================================
# 3. OPERACIONES ENTRE SERIES Y ALINEACIÓN DE DATOS (MÁSCARAS)
# =====================================================================
# Creamos dos series con algunos índices compartidos y otros diferentes
serie1 = pd.Series([10, 20, 30, 40], index=[0, 1, 2, 3])
serie2 = pd.Series([1, 2, 3], index=[0, 1, 4])

print("\n--- Serie 1 ---")
print(serie1)
print("\n--- Serie 2 ---")
print(serie2)

# Al sumar, los índices que no coinciden en ambas partes resultan en NaN
suma_directa = serie1 + serie2
print("\n--- Suma Directa (Nótese los NaN en índices 2, 3 y 4) ---")
print(suma_directa)

# OPERACIÓN DE PRE-FILTRADO/REEMPLAZO:
# Para evitar dejar valores NaN en producción, usamos una máscara con .isnull()
# y asignamos un valor por defecto (por ejemplo, 0) o usamos el método alternativo .add()
resultado_limpio = suma_directa.copy()
resultado_limpio[pd.isnull(resultado_limpio)] = 0

print("\n--- Resultado Limpiando NaNs (Reemplazados por 0) ---")
print(resultado_limpio)
```

# 📊 Manipulación de DataFrames en Pandas

## 1. Creación de Estructuras Bidimensionales
Los **DataFrames** permiten gestionar matrices de datos donde cada columna funciona como una `Series` independiente, pero perfectamente alineada bajo un índice común.

### 📌 Conceptos Clave:
* **Columnas:** Representan las variables del conjunto de datos.
* **Índices (Rows):** Identifican de manera única cada registro o fila.
* **Flexibilidad:** Permiten inicializar datos desde diccionarios, listas o matrices de NumPy.

```python
import pandas as pd

# =====================================================================
# 1. CREACIÓN BÁSICA DESDE UN DICCIONARIO
# =====================================================================
# Las claves del diccionario se convertirán automáticamente en columnas
datos_empleados = {
    "Nombre": ["Marisa", "Laura", "Manuel"],
    "Edad": [34, 29, 12]
}

# Creación por defecto (índices numéricos de 0 a N-1)
df_basico = pd.DataFrame(datos_empleados)
print("--- DataFrame Básico ---")
print(df_basico)


# =====================================================================
# 2. DATAFRAME CON ÍNDICES SEMÁNTICOS ASIGNADOS
# =====================================================================
# Asignamos etiquetas personalizadas a las filas usando el parámetro 'index'
df_con_indice = pd.DataFrame(datos_empleados, index=['a', 'b', 'c'])
print("\n--- DataFrame con Índices Semánticos ('a', 'b', 'c') ---")
print(df_con_indice)


# =====================================================================
# 3. CONTROL DE COLUMNAS Y VALORES AUSENTES (NaN)
# =====================================================================
# Si especificamos columnas que NO existen en el diccionario original,
# Pandas las creará automáticamente rellenándolas con valores nulos (NaN)
columnas_objetivo = ['Nacionalidad', 'Nombre', 'Edad', 'Profesion', 'Genero']

df_estructurado = pd.DataFrame(datos_empleados, columns=columnas_objetivo)
print("\n--- DataFrame con Columnas Expandidas (Manejo de NaN) ---")
print(df_estructurado)


# =====================================================================
# 4. ACCESO BÁSICO A COLUMNAS
# =====================================================================
# Método estándar y seguro usando corchetes (devuelve una estructura Series)
serie_nombres = df_estructurado['Nombre']
print("\n--- Extracción de la columna 'Nombre' (Tipo: Series) ---")
print(serie_nombres)
```

## 🛠️ 2. Modificación Estructural de DataFrames

Manipular la estructura de una matriz implica añadir nuevas variables, registrar nuevas observaciones o eliminar datos innecesarios.

### 📌 Conceptos Clave:
* [cite_start]**Asignación Directa:** Permite crear una columna nueva pasándole un valor escalar o una lista de la misma longitud[cite: 138, 139, 182, 188, 191].
* [cite_start]**Inserción de Filas:** Se utiliza `.loc[index]` para registrar una nueva fila suministrando todos los valores requeridos[cite: 194, 196].
* **Eliminación con `.drop()`:** Elimina filas o columnas. [cite_start]Por defecto opera en filas (`axis=0`); para columnas se debe especificar explícitamente `axis=1`[cite: 199, 202, 204].

```python
import pandas as pd

# Inicializamos el DataFrame base de la sección anterior para el ejercicio
diccionario = {
    "Nombre": ["Marisa", "Laura", "Manuel"],
    "Edad": [34, 29, 12]
}
frame = pd.DataFrame(diccionario, columns=['Nacionalidad', 'Nombre', 'Edad', 'Profesion', 'Direccion'])


# =====================================================================
# 1. AÑADIR COLUMNAS (VALOR ESCALAR Y LISTAS)
# =====================================================================

# Asignación de un valor escalar: Rellena toda la columna con el mismo dato
frame['Direccion'] = 'Desconocida'
print("--- Columna 'Direccion' con valor escalar ---")
print(frame)

# Asignación mediante una lista: Debe cumplir exactamente con la longitud del DataFrame
lista_direcciones = ['Rue 13 del Percebe, 13', 'Evergreen Terrace, 3', 'Av de los Rombos, 12']
frame['Direccion'] = lista_direcciones
print("\n--- Columna 'Direccion' actualizada con lista ---")
print(frame)


# =====================================================================
# 2. AÑADIR FILAS MEDIANTE .loc
# =====================================================================

# Para añadir una fila, pasamos una lista con los valores en el orden exacto de las columnas
user_2 = ['Alemania', 'Klaus', 20, 'none', 'Desconocida']
frame.loc[3] = user_2
print("\n--- DataFrame tras añadir la fila con índice 3 ---")
print(frame)


# =====================================================================
# 3. ELIMINAR FILAS Y COLUMNAS
# =====================================================================

# Eliminar una fila con .drop(): Genera una copia por defecto, a menos que uses inplace=True
# Eliminamos la fila con índice 2 (Manuel)
frame = frame.drop(2) 
print("\n--- Tras eliminar la fila con índice 2 ---")
print(frame)

# Eliminar una columna usando .drop() especificando axis=1
# Eliminamos la columna 'Nombre' modificando el objeto original directamente con inplace=True
frame.drop('Nombre', axis=1, inplace=True)
print("\n--- Tras eliminar la columna 'Nombre' (axis=1) ---")
print(frame)

# Eliminar otra columna usando la sentencia nativa 'del'
del frame['Profesion']
print("\n--- Tras eliminar la columna 'Profesion' con del ---")
print(frame)


# =====================================================================
# 4. OPERACIÓN EXTRA: MATRIZ TRASPUESTA
# =====================================================================
# Intercambia filas por columnas (T)
print("\n--- Vista de la Matriz Traspuesta (.T) ---")
print(frame.T)
```

## 🔍 3. Indexación Avanzada, Slicing y Selección de Datos

Esta sección aborda cómo extraer subconjuntos específicos de un `DataFrame` utilizando etiquetas semánticas o posiciones numéricas estrictas.

### 📌 Conceptos Clave:
* [cite_start]**`.iloc[row, col]`**: Selección basada exclusivamente en **posiciones enteras** (de 0 a N-1)[cite: 230, 244]. [cite_start]No incluye el límite superior en el slicing[cite: 232].
* [cite_start]**`.loc[row, col]`**: Selección basada en **etiquetas semánticas**[cite: 233, 246]. [cite_start]Incluye ambos extremos cuando se realiza un slicing[cite: 40, 233].
* [cite_start]**⚡ SettingWithCopyWarning**: Ocurre cuando modificas una porción del DataFrame obtenida por indexación encadenada[cite: 257]. [cite_start]Modificar una vista puede alterar el DataFrame original por referencia, por lo que se debe usar `.copy()` si se requiere un objeto independiente[cite: 251, 252, 260].

```python
import numpy as np
import pandas as pd

# =====================================================================
# 1. GENERACIÓN DEL DATAFRAME DE PRUEBA
# =====================================================================
# Construcción de un dataset a partir de una lista de diccionarios con datos meteorológicos
d1 = {'ciudad': 'Valencia', 'temperatura': 10, 'humedad': 90, 'co2': 10}
d2 = {'ciudad': 'Barcelona', 'temperatura': 8}
d3 = {'ciudad': 'Valencia', 'temperatura': 9}
d4 = {'ciudad': 'Madrid', 'temperatura': 10, 'humedad': 80}
d5 = {'ciudad': 'Sevilla', 'temperatura': 15, 'humedad': 50, 'co2': 6}
d6 = {'ciudad': 'Valencia', 'temperatura': 10, 'humedad': 90, 'co2': 10}

ls_data = [d1, d2, d3, d4, d5, d6]
df_data = pd.DataFrame(ls_data, index=list('abcdef'))

print("--- DataFrame Meteorológico ---")
print(df_data)


# =====================================================================
# 2. SELECCIÓN POSICIONAL CON .iloc
# =====================================================================

# Acceso a un valor concreto: Fila index 1, Columna index 1
val_especifico = df_data.iloc[1, 1]
print(f"\nValor en iloc[1,1] (Fila 1, Col 1): {val_especifico}")

# Slicing posicional: Primeras 3 filas y primeras 4 columnas (excluye límites superiores)
print("\n--- slicing posicional con .iloc[:3, :4] ---")
print(df_data.iloc[:3, :4])


# =====================================================================
# 3. SELECCIÓN SEMÁNTICA CON .loc
# =====================================================================

# Acceso explícito por etiquetas de fila y columna
val_semantico = df_data.loc['a', 'temperatura']
print(f"\nValor en loc['a', 'temperatura']: {val_semantico}")

# Slicing semántico: De la fila 'a' a la 'c', y columnas específicas (incluye extremos)
print("\n--- Slicing semántico con .loc[:'c', ['ciudad', 'co2']] ---")
# Nota: Ajustado el índice '02' erróneo del PDF al nombre real de la columna 'co2'
print(df_data.loc[:'c', ['ciudad', 'co2']])


# =====================================================================
# 4. EVITAR EL "SettingWithCopyWarning"
# =====================================================================

# Si extraemos una fila directa y modificamos un elemento, Pandas arrojará una advertencia
# porque se altera la estructura por referencia directa.
print("\n--- Intento de modificación directa (Peligro de Warning) ---")
serie_fila = df_data.loc['a']
# serie_fila.iloc[1] = 3000  # <-- Esto dispararía el SettingWithCopyWarning

# LA SOLUCIÓN CORRECTA: Forzar la clonación en memoria con .copy()
df_seguro = df_data.loc[['a']].copy()
df_seguro.iloc[0, 1] = 3000
print("DataFrame Original (Intacto):")
print(df_data.loc[['a']])
print("Copia Modificada de forma segura:")
print(df_seguro)
```

## 📁 4. Filtrado Avanzado, Carga y Guardado de Datos (CSV)

[cite_start]En esta sección consolidamos el uso de máscaras lógicas complejas en `DataFrames` y automatizamos el almacenamiento e importación de archivos hacia el disco duro utilizando la API de Pandas[cite: 327, 344].

### 📌 Conceptos Clave:
* [cite_start]**Operadores Lógicos Elementales:** * `&` (AND / Y lógico): Exige que ambas condiciones se cumplan simultáneamente[cite: 325].
  * [cite_start]`|` (OR / O lógico): Permite que se cumpla cualquiera de las condiciones dadas[cite: 324].
* [cite_start]**Paréntesis Obligatorios:** Al combinar criterios, cada bloque de condición debe estar obligatoriamente envuelto en `()` para evitar conflictos de precedencia de operadores en Python[cite: 325].
* [cite_start]**Manejo de Persistencia (`.to_csv` / `.read_csv`):** Permite escribir y leer estructuras tabulares controlando el separador de columnas (`sep`) y la codificación de datos nulos (`na_rep`)[cite: 328, 331, 332, 341].

```python
import os
import pandas as pd

# =====================================================================
# 1. GENERACIÓN DEL DATASET DE PERSONAJES (D&D)
# =====================================================================
d_and_d_characters = {
    'Name': ['bundenth', 'theorin', 'barlok'], 
    'Strength': [10, 12, 19], 
    'Wisdom': [20, 13, 6]
}
character_data = pd.DataFrame(d_and_d_characters, index=['a', 'b', 'c'])

print("--- Personajes Iniciales ---")
print(character_data)


# =====================================================================
# 2. FILTRADO COMPLEJO CON OPERADORES LÓGICOS
# =====================================================================

# Filtrar personajes con Strength > 11 Y Wisdom > 15 (AND lógico)
# Nota: Se corrigió la sintaxis inválida de corchetes del PDF original
condicion_and = (character_data['Strength'] > 11) & (character_data['Wisdom'] > 15)
print("\n--- Personajes con Fuerza > 11 Y Sabiduría > 15 ---")
print(character_data.loc[condicion_and])

# Filtrar personajes con Strength > 15 O Wisdom > 15 (OR lógico)
condicion_or = (character_data['Strength'] > 15) | (character_data['Wisdom'] > 15)
print("\n--- Personajes con Fuerza > 15 O Sabiduría > 15 ---")
print(character_data.loc[condicion_or])


# =====================================================================
# 3. ALMACENAMIENTO Y CARGA DE ARCHIVOS CSV
# =====================================================================

# Definición de ruta dinámica multiplataforma segura usando os.path
# Se crea un directorio temporal local llamado 'res' si no existe
if not os.path.exists("res"):
    os.makedirs("res")

ruta_archivo = os.path.join("res", "o_d_d_characters.csv")

# Exportar a CSV especificando punto y coma como delimitador
character_data.to_csv(ruta_archivo, sep=';', index=True)
print(f"\n[INFO] Archivo guardado con éxito en: {ruta_archivo}")

# Importar de regreso el CSV asignando la primera columna como índice real
loaded_data = pd.read_csv(ruta_archivo, sep=';', index_col=0)
print("\n--- Dataframe Recuperado desde el Disco ---")
print(loaded_data)
```

* [cite_start]**Normalización de Cargas:** Solventar problemas de archivos sin cabeceras (`header=None`) y delimitadores complejos (`sep='::'`)[cite: 363, 370].
* [cite_start]**Control y Limpieza de Errores de Muestreo:** Corregir datos demográficos corrompidos o inconsistentes mediante copias profundas explícitas para neutralizar el temido `SettingWithCopyWarning`[cite: 396, 403].

```python
import os
import zipfile
import urllib.request
import numpy as np
import pandas as pd

# =====================================================================
# 1. DESCARGA AUTOMATIZADA Y DESCOMPRESIÓN DEL DATASET
# =====================================================================

url = 'http://files.grouplens.org/datasets/movielens/ml-1m.zip'
directorio_destino = "res"
ruta_zip = os.path.join(directorio_destino, "ml-1m.zip")

# Crear el directorio base si no existe en el entorno local
if not os.path.exists(directorio_destino):
    os.makedirs(directorio_destino)

# Descargar el archivo zip de manera asíncrona/remota
if not os.path.exists(ruta_zip):
    print("[INFO] Descargando MovieLens Dataset (1M)...")
    urllib.request.urlretrieve(url, ruta_zip)
    print("[INFO] Descarga completada.")
else:
    print("[INFO] El archivo ZIP ya existe localmente.")

# Descompresión del archivo ZIP en el directorio de recursos
ruta_extraccion = os.path.join(directorio_destino)
if not os.path.exists(os.path.join(ruta_extraccion, "ml-1m")):
    print("[INFO] Extrayendo todos los archivos estructurados...")
    with zipfile.ZipFile(ruta_zip, 'r') as z:
        z.extractall(ruta_extraccion)
    print("[INFO] Extracción finalizada.")
else:
    print("[INFO] El dataset ya se encuentra descomprimido.")


# =====================================================================
# 2. INGESTA Y CONFIGURACIÓN ESTRUCTURAL DE DATOS (USERS.DAT)
# =====================================================================

ruta_users = os.path.join(directorio_destino, "ml-1m", "users.dat")

# Configuración profesional de lectura:
# El archivo original carece de cabecera y usa el delimitador especial '::'
# Forzamos engine='python' para que Pandas admita separadores multicarácter
users_dataset = pd.read_csv(
    ruta_users, 
    sep='::', 
    index_col=0,
    header=None, 
    names=['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code'],
    engine='python'
)

print("\n--- Muestra Aleatoria de Usuarios Extraída (.sample) ---")
print(users_dataset.sample(5))


# =====================================================================
# 3. AUDITORÍA Y PERFILAMIENTO SINTÁCTICO DE DATOS
# =====================================================================

print("\n--- Resumen Estadístico de Variables Numéricas ---")
print(users_dataset.describe())

# Contar cuántas observaciones pertenecen al género femenino ('F')
total_mujeres = len(users_dataset[users_dataset['Gender'] == 'F'])
print(f"\nNúmero total de usuarios registrados como Mujeres ('F'): {total_mujeres}")


# =====================================================================
# 4. TRATAMIENTO DE ANOMALÍAS (CORRECCIÓN SEGURA DE EDADES INCORRECTAS)
# =====================================================================

# El dataset codifica con '1' a los menores de edad/no especificados.
# Vamos a aislar ese segmento y neutralizarlo asignándole valores Nulos (NaN)
print(f"\nUsuarios con Edad codificada en 1: {len(users_dataset[users_dataset['Age'] == 1])}")

# SOLUCIÓN LIMPIA: Modificar directamente la porción usando .loc para evitar el Warning de copia
users_dataset.loc[users_dataset['Age'] == 1, 'Age'] = np.nan

# Mostrar que efectivamente se transformaron las filas críticas a NaN
print("\n--- Primeras filas con valores de Edad limpiados (NaN) ---")
print(users_dataset.loc[pd.isnull(users_dataset['Age'])].head(4))

# ELIMINACIÓN DE REGISTROS NULOS: Depurar el DataFrame original eliminando esos renglones
users_dataset.dropna(subset=['Age'], inplace=True)
print(f"\nEstructura final del dataset tras la limpieza profunda: {users_dataset.shape}")


# =====================================================================
# 5. PERSISTENCIA DE LOS DATOS PROCESADOS
# =====================================================================

ruta_output = os.path.join(directorio_destino, "ml-1m", "o_users_processed.csv")
# Almacenamos mapeando los NaNs como literales 'null' para compatibilidad
users_dataset.to_csv(ruta_output, sep=',', na_rep='null', index=True)
print(f"\n[INFO] Dataset final guardado de forma segura en: {ruta_output}")
```

## 🏁 Sección 6: Resolución Completa de Ejercicios de Evaluación

Llegamos a la última sección del PDF, donde el documento plantea un reto de análisis analítico sobre los archivos restantes del dataset MovieLens: `ratings.dat` y `movies.dat`. 

Como solicitaste, aquí desarrollamos las soluciones de manera explícita y profesional:
1. **Carga y Estructuración:** Ingesta de `ratings.dat` y `movies.dat` manejando delimitadores multicaracter (`::`).
2. **Auditoría de Reviews:** Identificar si existen usuarios huérfanos sin ninguna review asociada.
3. **Métrica de Distribución:** Calcular exactamente cuántos usuarios son considerados "usuarios de baja actividad" (menos de 30 reviews).

```python
import os
import pandas as pd

# Definición de rutas bases locales hacia los archivos estructurados
directorio_data = os.path.join("res", "ml-1m")
ruta_ratings = os.path.join(directorio_data, "ratings.dat")
ruta_movies = os.path.join(directorio_data, "movies.dat")


# =====================================================================
# 1. INGESTA PROFESIONAL DE DATOS
# =====================================================================

# Carga segura de Ratings (Evaluaciones)
ratings_dataset = pd.read_csv(
    ruta_ratings, 
    sep='::', 
    header=None, 
    names=['UserID', 'MovieID', 'Rating', 'Timestamp'],
    engine='python'
)

# Carga segura de Movies (Catálogo de películas)
# Nota: Forzamos encoding='ISO-8859-1' o 'latin-1' debido a caracteres especiales en los títulos
movies_dataset = pd.read_csv(
    ruta_movies, 
    sep='::', 
    header=None, 
    names=['MovieID', 'Title', 'Genres'],
    engine='python',
    encoding='ISO-8859-1'
)

print("--- Dimensiones de las Tablas Cargadas ---")
print(f"Dataset de Evaluaciones (Filas, Columnas): {ratings_dataset.shape}")
print(f"Dataset de Películas (Filas, Columnas): {movies_dataset.shape}")


# =====================================================================
# 2. RESOLUCIÓN EJERCICIO 1: ¿HAY USUARIOS SIN NINGUNA REVIEW?
# =====================================================================

# Obtenemos cuántos IDs únicos registraron al menos una evaluación
usuarios_con_review = ratings_dataset['UserID'].nunique()

# Para cruzar datos con el archivo de usuarios del bloque anterior:
# Si asumimos que los IDs de usuarios son secuenciales y van del 1 al máximo ID:
max_user_id = ratings_dataset['UserID'].max()

print("\n--- Auditoría de Registros Activos ---")
print(f"Cantidad de usuarios únicos que SÍ tienen reviews: {usuarios_con_review}")
print(f"ID de usuario máximo detectado en la matriz: {max_user_id}")

if usuarios_con_review == max_user_id:
    print("Resultado: NO existen usuarios vacíos. Todos los IDs intermedios registran actividad.")
else:
    print(f"Resultado: Existen {max_user_id - usuarios_con_review} IDs sin registros de reviews.")


# =====================================================================
# 3. RESOLUCIÓN EJERCICIO 2: ¿CUÁNTOS TIENEN MENOS DE 30 REVIEWS?
# =====================================================================

# Contabilizar la cantidad de apariciones de cada UserID (frecuencia de reviews)
reviews_por_usuario = ratings_dataset['UserID'].value_counts()

# Filtrar la Serie resultante aplicando una máscara lógica indexada
usuarios_baja_actividad = reviews_por_usuario[reviews_por_usuario < 30]

print("\n--- Análisis de Distribución de Frecuencias ---")
print(f"Número de usuarios que tienen menos de 30 reviews: {len(usuarios_baja_actividad)}")

# Muestra descriptiva de los primeros 5 usuarios en este umbral crítico
print("\nMuestra de IDs de usuario junto a su conteo total de reviews (< 30):")
print(usuarios_baja_actividad.head(5))
```
