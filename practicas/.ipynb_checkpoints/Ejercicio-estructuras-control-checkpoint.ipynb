## 1. Estructuras de Control

### 1.1 Selección (sentencias condicionales)

- Selección de una de varias alternativas en base a alguna condición.
- La indentación estructura el código.
- Es importante el símbolo `:`

````python
# Asigno el valor uno a la variable x
x = int(input("Ingrese el valor de x"))
# Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
````

````python
x = 3
y = [1, 2, 3]
print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
````

Se pueden introducir más "ramas" en sentencias condicionales a través de la palabra clave `elif`.

````python
x = -3
if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo')
else:
    print('Valor negativo')
````

Anidamiento de sentencias condicionales.

````python
val = 120
if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')
else:
    print('Valor negativo')
````

## 2. Switch

````python
a = -6
match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        print('negative')
````

````python
a = 'l'  # week day
match a:
    case 'L':
        print("Lunes")
    case 'M':
        print("Martes")
    case 'X':
        print("Miercoles")
    case _:
        print("Wrong Day")
````

### Expresión ternaria

- Estructura if-else condensada en una línea.
- Se recomienda usar solo en casos sencillos.

````python
# Ejemplo: cálculo del valor absoluto de un número.
x = -3
# Expresión ternaria: si x es mayor o igual a 0, resultado = x; si no, resultado = -x
resultado = x if x >= 0 else -x
print(resultado)
````

````python
val = -3
if val >= 0:
    resultado = val
else:
    resultado = -val
print(resultado)
````

### Valor booleano

- Todos los objetos en Python tienen inherentemente un valor booleano: `True` o `False`.
- Cualquier número distinto de 0 o cualquier objeto no vacío tiene valor `True`.
- El número 0, objetos vacíos y el objeto especial `None` tienen valor `False`.

````python
a = -10  # True
b = None  # False
if a:
    print("a es True")
if not b:
    print("b es False")
````

````python
c = [1, 2, 3]
# Reasigno c a None para simular una lista "vacía"/nula y comprobar su valor booleano
c = None
if c:
    print("Lista no vacía")
else:
    print("Lista vacía")
````

## 3. Iteración (bucles)

- Repetición de un bloque de código.
- La terminación del bucle depende del tipo de bucle.
- Dos tipos: `while` y `for`.

### Bucle `while`

- Repetición de un bloque de código hasta que se deje de cumplir una expresión (es decir, hasta que una condición evalúe a `False`).
- Si la condición evalúa a `False` desde el principio, el bloque de código nunca se ejecuta.
- Cuidado con los bucles infinitos.

Formato general:
```python
while test:      # Mientras se cumple la condición
    statements    # Instrucciones a ejecutar
```

````python
# Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]
while indice < 3:
    print(numeros[indice])
    indice = indice + 1
````

````python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0
while indice < len(numeros):
    # Reviso cada número de la lista; si es menor a 10 lo muestro y sumo al contador
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1
print('Contador:', contador)
print('Indice:', indice)
````

````python
nombre = 'Pablo'
while nombre:  # Mientras 'nombre' no sea vacío
    print(nombre)
    nombre = nombre[1:]
````

````python
# Aunque el comentario original dice "bucle infinito", en realidad SÍ termina
# porque la condición i < 10 eventualmente se vuelve False
i = 0
while i < 10:
    print(i)
    i += 1
````

### Bucle `for`

- Permite recorrer los items de una secuencia o un objeto iterable.
- Funciona en strings, listas, tuplas, etc.

Formato general:
```python
for item in objeto:   # Asigna los items del objeto a la variable item en cada iteración
    statements         # Instrucciones a ejecutar
```

````python
peliculas = ['Matrix', 'The purge', 'Avatar', 'Star Wars']
for pelicula in peliculas:
    print(pelicula)
````

También se usa para iterar un número preestablecido de veces (counted loops):

````python
for i in range(10):
    print(i)
````

````python
nombre = 'Pablo'
for i in range(len(nombre)):
    print(nombre[i])
````

````python
for letra in nombre:
    print(letra)
````

````python
# Nota: esta celda genera un error a propósito.
# La primera tupla (1, 2, 4) tiene 3 elementos, pero (3, 4) solo tiene 2,
# y estamos intentando desempaquetar cada tupla en 3 variables (a, b, c)
tuplas = [(1, 2, 4), (3, 4), (5, 6)]
for a, b, c in tuplas:
    print(a, b, c)
````

### Iteración de diccionarios

Se iteran las claves:

````python
diccionario = {'a': 1, 'b': 2, 'c': 3}
for key in diccionario:
    print(f"{key} => {diccionario[key]}")
````

Para iterar los pares (clave-valor) o únicamente los valores, se deben usar los métodos `items` y `values`.

````python
for key, value in diccionario.items():
    print(f"{key} => {value}")
````

````python
for value in diccionario.values():
    print(value)
````

La variable `item` en la cabecera del `for` puede ser cualquier expresión que sea válida como parte izquierda de una asignación convencional.

````python
for a, b, c in [(1, 2, 3), (4, 5, 6)]:
    print(a, b, c)
````

### 3.1 Las sentencias break, continue y else

**Break y continue**
- Solo tienen sentido dentro de bucles.
- `break` permite terminar el bucle por completo.
- `continue` permite saltar a la siguiente iteración, continuando con el bucle.
- Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sentencias condicionales (`if`).

````python
# Ejemplo: break
# Busco un rating específico en la lista; cuando lo encuentro, muestro "Found" y detengo el bucle
rating_to_find = 4.2
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]
for rating in movie_ratings:
    print(rating)
    if rating == rating_to_find:
        print("Found")
        break
print(rating)
````

````python
for i in range(10):
    if i % 2 == 0:
        continue  # Si el número es par, salta a la siguiente iteración
    print(f'Odd number {i}')
````

Observa cómo la sentencia `continue` te puede ayudar a reducir el número de niveles de anidamiento. Sin `continue`, el anterior ejemplo sería:

````python
for i in range(10):
    if i % 2 != 0:
        print('Numero impar:', end=' ')
        print(i)
````

````python
# Nota: el 'break' aquí solo detiene el bucle interno (el de 'a'), no el externo (el de 'i')
for i in range(5):
    for a in range(2):
        print(f"{i} - {a}")
        if i == 3 and a == 0:
            break
````

### Else

- Los bucles pueden tener una sentencia `else`.
- Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros lenguajes.
- Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de un `break`.

````python
lista = [60, 60, 30, 60]
element_to_find = 60
for element in lista:
    if element == element_to_find:
        print("found")
        break
else:
    print("not found")
````

````python
# Misma lógica que la celda anterior, pero usando una variable "found" como bandera
# en vez de aprovechar el 'else' del bucle for
lista = [10, 30, 50, 40]
element_to_find = 30
found = False
for element in lista:
    if element == element_to_find:
        found = True
        break
if found:
    print("dato encontrado")
else:
    print("not found")
````

## 4. List Comprehensions

- Permiten construir listas a través de la ejecución repetida (sentencia `for`) de una expresión para cada item de un objeto iterable.
- Van entre `[` y `]`. Esto es indicativo de que estamos construyendo una lista.

Sintaxis:
```python
[<expression> for <item> in <iterable>]
```

Ejemplo: repetir los caracteres de un string.

````python
[char*3 for char in "Enrique"]
````

````python
lista = []
for char in 'Enrique':
    lista.append(char)
print(lista)
````
